```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Read the data
df = pd.read_csv("01_Power Failure.csv", delimiter=';')
data = pd.DataFrame(df)

# Create a figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces for I_L1real
fig.add_trace(
    go.Scatter(x=data['Timestamp'], y=data['I_L1real'], mode='lines', name='I_L1real', line=dict(color='#d62728')),
    secondary_y=False,
)

# Add traces for U_L1real
fig.add_trace(
    go.Scatter(x=data['Timestamp'], y=data['U_L1real'], mode='lines', name='U_L1real', line=dict(color='#1f77b4')),
    secondary_y=True,
)

# Update layout
fig.update_layout(
    title='Power Failure',
    xaxis_title='Timestamp',
    yaxis_title='Current (mA)',
    legend_title='Signals'
)

# Set y-axes titles
fig.update_yaxes(title_text="Current (mA)", range=[-0.5, 0.5], secondary_y=False)
fig.update_yaxes(title_text="Voltage (V)", range=[-2, 2], secondary_y=True)

# Show figure
fig.show()
```
